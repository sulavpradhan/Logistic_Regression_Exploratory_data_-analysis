```python
import pandas as pd

def perform_eda(file_path):
    # 1. We map the variables to their exact column indices (0-indexed)
    # Age (1), Education (3), Nscore (6), Escore (7), Oscore (8), 
    # Ascore (9), Cscore (10), Impulsive (11), SS (12)
    target_indices = [1, 3, 6, 7, 8, 9, 10, 11, 12]
    
    # 2. Assign the exact names from the assignment prompt
    target_names = [
        'Age', 'Education', 'Nscore', 'Escore', 
        'Oscore', 'Ascore', 'Cscore', 'Impulsive', 'SS'
    ]
    
    # 3. Load the data
    # We use header=None because the raw data has no title row
    # We use usecols to save memory and only extract what the assignment asks for
    df = pd.read_csv(file_path, header=None, usecols=target_indices)
    df.columns = target_names
    
    # 4. Generate the 8 required statistics
    # describe() automatically outputs: count, mean, std, min, 25%, 50%, 75%, max
    summary_stats = df.describe().round(4) # Rounding to 4 decimal places for readability
    
    return summary_stats

# --- Execute the script ---
if __name__ == "__main__":
    # Replace with the actual path to your downloaded file
    file_path = "./drug_consumption/drug_consumption.data"
    
    try:
        stats = perform_eda(file_path)
        print("Exploratory Data Analysis - Summary Statistics:\n")
        print(stats.to_string())
    except FileNotFoundError:
        print(f"Error: Could not find '{file_path}'. Please ensure it is in the same directory.")
```

```python
import math

def calculate_summary(data):
    # count, mean, standard deviation, minimum, 25% percentile, 50% percentile, 75% percentile, and maximum
    count = len(data)

    if count == 0:
        return None

    mean = sum(data) / count

    std_dev = math.sqrt(sum([(x - mean) ** 2 for x in data]) / (count - 1))

    min_val = min(data)
    max_val = max(data)

    sorted_data = sorted(data)
    rank_25 = count * 0.25
    rank_50 = count * 0.5
    rank_75 = count * 0.75

    percentile_25 = get_percentile(sorted_data, rank_25)
    percentile_50 = get_percentile(sorted_data, rank_50)
    percentile_75 = get_percentile(sorted_data, rank_75)

    return ({
        'count': count,
        'mean': mean,
        'std_dev': std_dev,
        'min': min_val,
        'max': max_val,
        'percentile_25': percentile_25,
        'percentile_50': percentile_50,
        'percentile_75': percentile_75,
    })
```
